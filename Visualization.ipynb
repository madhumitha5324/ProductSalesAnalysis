```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('sales_dataset_comprehensive.csv', parse_dates=['Date'])
```

### 1. Total sales by month


```python
monthly_sales = df.groupby(['Year', 'Month_Name'])['Total_Amount'].sum().reset_index()
plt.figure(figsize=(12,6))
sns.lineplot(data=monthly_sales, x='Month_Name', y='Total_Amount', hue='Year', marker='o')
plt.title('Total Sales by Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 2. Sales by product category


```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Category', y='Total_Amount', estimator=sum, ci=None)
plt.title('Total Sales by Product Category')
plt.tight_layout()
plt.show()
```

### 3. Sales distribution by region


```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Region', y='Total_Amount')
plt.title('Sales Distribution by Region')
plt.tight_layout()
plt.show()
```

### 4. Top 10 products by total sales


```python
top_products = df.groupby('Product')['Total_Amount'].sum().nlargest(10).reset_index()
plt.figure(figsize=(10,6))
sns.barplot(data=top_products, x='Total_Amount', y='Product')
plt.title('Top 10 Products by Total Sales')
plt.tight_layout()
plt.show()
```

### 5. Sales trend by order priority


```python
priority_sales = df.groupby(['Order_Priority', 'Month_Name'])['Total_Amount'].sum().reset_index()
plt.figure(figsize=(12,6))
sns.lineplot(data=priority_sales, x='Month_Name', y='Total_Amount', hue='Order_Priority', marker='o')
plt.title('Sales Trend by Order Priority')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 6. Correlation heatmap for numeric features


```python
plt.figure(figsize=(10,8))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
