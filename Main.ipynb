```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading the CSV file

```python
df = pd.read_csv("product_data.csv")
```

 Exploration

```python
print("Shape:", df.shape)
print(df.dtypes)
print(df.head(3))
```

Drop duplicate rows

```python
df.drop_duplicates(inplace=True)
df
```

Drop rows with critical missing values

```python
df.dropna(subset=['actual_price', 'selling_price', 'average_rating', 'category', 'brand'], inplace=True)
```

Fill missing ratings

```python
df['average_rating'] = pd.to_numeric(df['average_rating'], errors='coerce')  
df.fillna({'average_rating': df['average_rating'].mean()}, inplace=True)

df
```

Create a New Column

```python
df['estimated_units'] = df['UnitsSold'] if 'UnitsSold' in df.columns else 100  # fallback
df['revenue'] = df['selling_price'] * df['estimated_units']
df
```

Filtering

```python
filtered_df = df[(df['category'] == 'Clothing and Accessories') & (df['average_rating'] >= 4.0)]
print("Filtered products:\n", filtered_df[['title', 'average_rating']])
```

Sorting

```python
top_revenue = df.sort_values(by='revenue', ascending=False).head(5)
print("Top 5 by Revenue:\n", top_revenue[['title', 'revenue']])
```

Grouping - Revenue by Brand

```python
revenue_by_brand = df.groupby('brand')['revenue'].sum().sort_values(ascending=False)
print("Revenue by Brand:\n", revenue_by_brand)
```

Visualization 

```python
plt.figure(figsize=(8, 4))
plt.plot(df['revenue'])
plt.title("Line Plot: Revenue Over Products")
plt.xlabel("Product Index")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x='category', y='revenue', estimator=sum)
plt.title("Revenue by Category")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
