### GROUPING DATA

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('sales_dataset_comprehensive.csv')
```

```python
df.groupby('Region').agg({
    'Total_Amount': ['count', 'sum', 'mean', 'std',],
    'Quantity': 'sum',
    'Customer_Satisfaction': 'mean'
}).round(3)
```

```python
df.groupby(['Product', 'Region']).agg({
    'Total_Amount': ['count', 'sum', 'mean'],
    'Customer_Satisfaction': 'mean'
}).round(2)
```
