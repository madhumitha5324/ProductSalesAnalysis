```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('sales_dataset_comprehensive.csv')
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

#### Data Manipulation

```python
pd.Timestamp.now() - df['Date']
```

```python
df['CustomerRelation'] = pd.cut(
    df['Date'],
    bins = 3,
    labels = ['New', 'Returning', 'Loyal']
)
```

```python
# Adding new columns
df['Days_Since_Purchase'] = (pd.Timestamp.now() - df['Date']).dt.days
df['Revenue_Category'] = pd.cut(df['Total_Amount'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
df['Bulk_Order'] = df['Quantity'].apply(lambda x: 'Yes' if x >= 10 else 'No')
```

```python
df[['Subtotal','Total_Amount']].head()
```

```python
df_cleaned = df.drop(['Days_Since_Purchase'], axis=1)
df_cleaned.shape
```
